```python
# Import required libraries
import pandas as pd
import random
import ipywidgets as widgets
from IPython.display import display

# Load CSV data
data = pd.read_csv('extracted_text.csv')

# Define the list of positions
# Define the list of positions
positions = ['Pitcher', 'Catcher', '1. Baseman', '2. Baseman', '3. Baseman', 'Shortstop', 'Leftfield','Centerfield', 'Rightfield']
# Create a set of checkboxes for positions
checkboxes = [widgets.Checkbox(value=False, description=pos) for pos in positions]

# Display the checkboxes
checkboxes_box = widgets.VBox(checkboxes)
display(checkboxes_box)

def get_selected_positions():
    return [cb.description for cb in checkboxes if cb.value]

def get_random_situation(selected_positions):
    if not selected_positions:
        return None, "Please select at least one position."

    # Filter the data based on selected positions
    filtered_data = data[['Situation', 'Outs', 'Runners'] + selected_positions]
    random_situation = filtered_data.sample(1).iloc[0]

    situation = random_situation['Situation']
    outs = random_situation['Outs']
    runners = random_situation['Runners']
    answer = random_situation[selected_positions]

    return (situation, outs, runners), answer

# Button to trigger the selection
button = widgets.Button(description="Get Random Situation")
output = widgets.Output()

def on_button_clicked(b):
    with output:
        output.clear_output()
        selected_positions = get_selected_positions()
        if not selected_positions:
            print("Please select at least one position.")
        else:
            situation, answer = get_random_situation(selected_positions)
            print("Situation:", situation)
            print("Answer:", answer)

button.on_click(on_button_clicked)

# Arrange widgets in a grid layout
app_layout = widgets.VBox([checkboxes_box, button, output])
display(app_layout)

# Create a set of checkboxes for positions
checkboxes = [widgets.Checkbox(value=False, description=pos) for pos in positions]

# Display the checkboxes
checkboxes_box = widgets.VBox(checkboxes)
display(checkboxes_box)

def get_selected_positions():
    return [cb.description for cb in checkboxes if cb.value]

def get_random_situation(selected_positions):
    if not selected_positions:
        return None, "Please select at least one position."

    # Filter the data based on selected positions
    filtered_data = data[['Situation', 'Outs', 'Runners'] + selected_positions]
    random_situation = filtered_data.sample(1).iloc[0]

    situation = random_situation['Situation']
    outs = random_situation['Outs']
    runners = random_situation['Runners']
    answer = random_situation[selected_positions]

    return (situation, outs, runners), answer

# Button to trigger the selection
button = widgets.Button(description="Get Random Situation")
output = widgets.Output()

def on_button_clicked(b):
    with output:
        output.clear_output()
        selected_positions = get_selected_positions()
        if not selected_positions:
            print("Please select at least one position.")
        else:
            situation, answer = get_random_situation(selected_positions)
            print("Situation:", situation)
            print("Answer:", answer)

button.on_click(on_button_clicked)
display(button, output)
```
